```python
import os
import pandas as pd
from tqdm import tqdm
import csv
```

```python
df_1 = pd.read_csv("./metadata/before_after/2019-dec-02.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_2 = pd.read_csv("./metadata/before_after/2019-dec-03.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_3 = pd.read_csv("./metadata/before_after/2019-dec-04.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_4 = pd.read_csv("./metadata/before_after/2019-dec-05.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_5 = pd.read_csv("./metadata/before_after/2019-dec-06.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_6 = pd.read_csv("./metadata/before_after/2019-dec-07.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_7 = pd.read_csv("./metadata/before_after/2019-dec-08.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7])
df = df[df["sub_category"] != "Fitness and Recreational Sports Centers"]
df = df.groupby("sub_category")[["Fitness_First", "Fitness_Later"]].sum()
df["Fitness_First_Percentage"] = df["Fitness_First"] / (df["Fitness_First"] + df["Fitness_Later"])
df["Fitness_Later_Percentage"] = df["Fitness_Later"] / (df["Fitness_First"] + df["Fitness_Later"])
df_removed_smaller = df[df["Fitness_First"] + df["Fitness_Later"] > 50000]
df_removed_smaller["rank"] = df_removed_smaller["Fitness_First_Percentage"].rank(ascending=False)
unique_1 = df_removed_smaller.index.unique()
```

```python
df_1 = pd.read_csv("./metadata/before_after/2019-oct-07.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_2 = pd.read_csv("./metadata/before_after/2019-oct-08.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_3 = pd.read_csv("./metadata/before_after/2019-oct-09.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_4 = pd.read_csv("./metadata/before_after/2019-oct-10.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_5 = pd.read_csv("./metadata/before_after/2019-oct-11.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_6 = pd.read_csv("./metadata/before_after/2019-oct-12.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df_7 = pd.read_csv("./metadata/before_after/2019-oct-13.csv", header=None, names = ["sub_category", "Fitness_First", "Fitness_Later"])
df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7])
df = df[df["sub_category"] != "Fitness and Recreational Sports Centers"]
df = df.groupby("sub_category")[["Fitness_First", "Fitness_Later"]].sum()
df["Fitness_First_Percentage"] = df["Fitness_First"] / (df["Fitness_First"] + df["Fitness_Later"])
df["Fitness_Later_Percentage"] = df["Fitness_Later"] / (df["Fitness_First"] + df["Fitness_Later"])
df_removed_smaller_2 = df[df["Fitness_First"] + df["Fitness_Later"] > 50000]
df_removed_smaller_2["rank"] = df_removed_smaller_2["Fitness_First_Percentage"].rank(ascending=False)
unique_2 = df_removed_smaller_2.index.unique()
```

```python
common = list(set(unique_1).intersection(unique_2))
print(len(common)) # Number of subcategories confident in both months
```

```python
df_removed_smaller = df_removed_smaller[df_removed_smaller.index.isin(common)]
df_removed_smaller_2 = df_removed_smaller_2[df_removed_smaller_2.index.isin(common)]
```

```python
# December
df_removed_smaller["rank"] = df_removed_smaller["Fitness_First_Percentage"].rank(ascending=False)
print(df_removed_smaller.sort_values(["Fitness_First_Percentage"], ascending=False))
```

```python
# October
df_removed_smaller_2["rank"] = df_removed_smaller_2["Fitness_First_Percentage"].rank(ascending=False)
print(df_removed_smaller_2.sort_values(["Fitness_First_Percentage"], ascending=False))
```
