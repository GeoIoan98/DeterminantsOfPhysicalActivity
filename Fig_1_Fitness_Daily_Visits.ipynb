```python
import os 
import pandas
from tqdm import tqdm
import plotly.express as px
from datetime import datetime
```

```python
files = os.listdir("../daily_visits")
files = [os.path.join(os.getcwd(), "../daily_visits", file) for file in files]
```

```python
visits = pandas.DataFrame()
for f in tqdm(sorted(files)):
    visits = pandas.concat([visits, pandas.read_csv(f)])
visits = visits[visits["location"] == "Planet Fitness"] # Keep only Planet Fitness
visits.reset_index(inplace=True, drop=True)
visits.drop(visits.filter(regex="Unname"),axis=1, inplace=True) # drop the first unnamed column
visits
```

```python
# convert name of month into number
month_conv = {"jan": 1, "feb": 2, "mar": 3, "apr": 4, "may": 5, "jun": 6, "jul": 7, "aug": 8, "sep": 9,
              "oct": 10, "nov": 11, "dec": 12}
# create the datetime dates
dates = []
for i, r in visits.iterrows():
    date = datetime(visits["year"].iloc[i], month_conv[visits["month"].iloc[i]], visits["day"].iloc[i])
    date = datetime.date(date)
    dates += [date]
visits["date"] = dates
visits
```

```python
# Sort rows by date
visits.sort_values(by = "date", inplace=True)
visits.reset_index(inplace=True, drop=True)
visits
```

```python
fig = px.line(visits, x='date', y="total_visits", labels={
                     "total_visits": "planet fitness visits"
                 }, width=800, height=300)
fig.update_layout(xaxis_title=None)
fig.show()
```

```python
fig.write_image("PlanetFitnessYearlyVisits.pdf")
```
