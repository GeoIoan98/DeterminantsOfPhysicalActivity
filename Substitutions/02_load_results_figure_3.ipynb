```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick
```

```python
base = "./metadata/subs_figures/dec-02-05-final-nonoverlapping.pkl"
with open(base, 'rb') as f:
    group_1_day_visits, group_1_day_total, group_2_days_visits, \
    group_2_days_total, group_3_days_visits, group_3_days_total = pickle.load(f)
```

```python
sub_category = "Full-Service Restaurants"
plt.figure()
plt.xticks([0,1,2,3], ["02", "03", "04", "05"])
plt.title("Full-Service Restaurants")

# Handle group 1
xs = [2,3]
y_1 = group_1_day_visits[sub_category][0]/group_1_day_total[0]
y_2 = group_1_day_visits[sub_category][1]/group_1_day_total[1]
ys = [y_1, y_2]
plt.plot(xs, ys, label = "Exercised a single day before resting")
ci_1 = 1.96 * ( (y_1*(1-y_1))/group_1_day_total[0] )**0.5
ci_2 = 1.96 * ( (y_2*(1-y_2))/group_1_day_total[1] )**0.5
plt.fill_between(xs, [y_1 + ci_1, y_2 + ci_2], [y_1 - ci_1, y_2 - ci_2], alpha = .3)

# Handle group 2
xs = [1,2,3]
y_1 = group_2_days_visits[sub_category][0]/group_2_days_total[0]
y_2 = group_2_days_visits[sub_category][1]/group_2_days_total[1]
y_3 = group_2_days_visits[sub_category][2]/group_2_days_total[2]
ys = [y_1, y_2, y_3]
plt.plot(xs, ys, label = "Exercised for two days before resting")
ci_1 = 1.96 * ( (y_1*(1-y_1))/group_2_days_total[0] )**0.5
ci_2 = 1.96 * ( (y_2*(1-y_2))/group_2_days_total[1] )**0.5
ci_3 = 1.96 * ( (y_3*(1-y_3))/group_2_days_total[2] )**0.5
plt.fill_between(xs, [y_1 + ci_1, y_2 + ci_2, y_3 + ci_3], [y_1 - ci_1, y_2 - ci_2, y_3 - ci_3], alpha = .3)

# Handle group 3
xs = [0,1,2,3]
y_1 = group_3_days_visits[sub_category][0]/group_3_days_total[0]
y_2 = group_3_days_visits[sub_category][1]/group_3_days_total[1]
y_3 = group_3_days_visits[sub_category][2]/group_3_days_total[2]
y_4 = group_3_days_visits[sub_category][3]/group_3_days_total[3]
ys = [y_1, y_2, y_3, y_4]

plt.plot(xs, ys, label = "Exercised for at least three days before resting")

plt.xlabel("Date: December, 2019")
plt.ylabel("Percentage of Visits")

ci_1 = 1.96 * ( (y_1*(1-y_1))/group_3_days_total[0] )**0.5
ci_2 = 1.96 * ( (y_2*(1-y_2))/group_3_days_total[1] )**0.5
ci_3 = 1.96 * ( (y_3*(1-y_3))/group_3_days_total[2] )**0.5
ci_4 = 1.96 * ( (y_4*(1-y_4))/group_3_days_total[3] )**0.5
plt.fill_between(xs, [y_1 + ci_1, y_2 + ci_2, y_3 + ci_3, y_4 + ci_4], [y_1 - ci_1, y_2 - ci_2, y_3 - ci_3, y_4 - ci_4], alpha = .3)

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
plt.legend()
plt.tight_layout()
```
